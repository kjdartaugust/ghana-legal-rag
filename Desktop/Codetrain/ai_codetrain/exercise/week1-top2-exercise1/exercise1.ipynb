```python
# Prices for: Maize, Rice, Yam, Cassava, Tomatoes, Onions
crop_prices = [45.5, 120.0, 35.0, 15.0, 60.0, 22.5]

# Print only the first 3 prices
print(f"First 3 prices: {crop_prices[:3]}")
```

```python
# Get last price using negative indexing
print(f"Last price: {crop_prices[-1]}")

# Reverse the list
reversed_prices = crop_prices[::-1]
print(f"Reversed list: {reversed_prices}")

# Calculations
avg_price = sum(crop_prices) / len(crop_prices)
print(f"Max: {max(crop_prices)}, Min: {min(crop_prices)}, Avg: {avg_price:.2f}")
```
